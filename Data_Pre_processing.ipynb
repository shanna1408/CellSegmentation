```python
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
import numpy as np

# Simulating 25 example embeddings of size 256
np.random.seed(42)  # For reproducibility
example_embeddings = np.random.uniform(-1, 1, (25, 256))

# Example: View first embedding
print("First Embedding:", example_embeddings[0])
```

```python
import numpy as np

mean = np.mean(example_embeddings)
std = np.std(example_embeddings)

normalized_embeddings = (example_embeddings - mean) / std

print("First Normalized Embedding:", normalized_embeddings[0])
```

```python
n_components = 2
pca = PCA(n_components=n_components)
reduced_embeddings = pca.fit_transform(normalized_embeddings)

# View the first few reduced embeddings
print(f"Reduced Embeddings (first 5 samples): \n{reduced_embeddings[:5]}")
```

```python
plt.scatter(reduced_embeddings[:, 0], reduced_embeddings[:, 1], c='blue', s=50)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of Example Embeddings (2D)')
plt.show()
```
